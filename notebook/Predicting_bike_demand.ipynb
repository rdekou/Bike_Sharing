```python
num_folds = 3

cv= KFold(n_splits= num_folds)
```

##### Casual lightgbm model

```python
def hyperopt_train_test(params):


    print("Training with params: ")

    print(params)

    num_round = int(params['n_estimators'])

    del params['n_estimators']

    results=np.repeat(0.0,num_folds)

    
    i = 0
    
    for train_index, test_index in  cv.split(train_features): 

       evals_results = {}
    
       dtrain = lgb.Dataset(train_features.iloc[train_index], label=y_train_c.iloc[train_index])
       dvalid =  lgb.Dataset(train_features.iloc[test_index], label=y_train_c.iloc[test_index])
    
       bst1 = lgb.train(params, 

                     dtrain, 

                     valid_sets=[dtrain ,dvalid], 

                     valid_names=['train', 'valid'], 

                     evals_result=evals_results, 

                     num_boost_round= num_round,

                     early_stopping_rounds=20,

                     verbose_eval=10 

                       )

        

       results[i] = evals_results['valid']['rmse'][bst1.best_iteration-1]
        
       i = i +1


        # TODO: Add the importance for the selected features

    

    score = np.mean(results)

    print("\tScore {0}\n\n".format(score))

    # The score function should return the loss (1-score)

    # since the optimize function looks for the minimum

    # loss = 1 - score

    return score #{'loss': loss, 'status': STATUS_OK}



spacelgb = {

        'n_estimators': hp.quniform('n_estimators', 100, 1000, 1),

        'eta': hp.quniform('eta', 0.025, 0.5, 0.025),

        'max_depth':  hp.choice('max_depth', np.arange(3, 14, dtype=int)),

        'min_child_weight': hp.quniform('min_child_weight', 2, 10, 1),

        'subsample': hp.quniform('subsample', 0.5, 1, 0.05),


        'colsample_bytree': hp.quniform('colsample_bytree', 0.5, 1, 0.05),

        'metric': 'rmse',
 

        'nthread': 4,

        'boosting_type': hp.choice('boosting_type', [ 'gbdt']),

        'tree_method': 'exact',

        'silent': 1,

        'seed': 1 # random_state

    }



best = 1000000

def f(params):

    global best
    
    acc = hyperopt_train_test(params)

    if acc < best:

      best = acc
    
      print( 'new best:', best, params)


    return {'loss': acc, 'status': STATUS_OK}

```
